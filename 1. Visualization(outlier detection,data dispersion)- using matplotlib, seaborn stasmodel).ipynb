```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# for identiication of color variation
sns.set(color_codes = True)

#use of plot method present in the pyplot library from matplotlib
plt.plot(np.arange(20))
```

```
# for plotting one dimensional data with respect to its indices
x = np.random.randn(15)
plt.plot(x)
plt.title('Plot using matplotlib')
plt.xlabel('Index')
plt.ylabel('Random Value')
```

```
# plot 2D data
x = [1, 2, 4, 7]
y = [4, 2, 6, 2]
plt.plot(x, y)
```

```
# here a1 to a6 represent attributes
data = np.random.randint(10, size=(10, 6))
df = pd.DataFrame(data, columns=['a1', 'a2', 'a3', 'a4', 'a5', 'a6'])
df
```

```
# Boxplot using pandas plot.box function
# It hass got various parameters for better representation of data
#
df.plot.box(grid = True, patch_artist = True)
```

```
# Boxplot using pandas boxplot function
# Here we can select which attributes have to be plotted
#we can even rotate rthe nameof attributes
#
df.boxplot(grid = True, patch_artist = True, column = ['a2', 'a6'], rot = 90
```
