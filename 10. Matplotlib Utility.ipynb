```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import random

days=[]
min_t=[]
max_t=[]
avg_t=[]
temp = []

for i in range(50):
    days.append(i)
    tmax = random.randint(25,40)
    tmin = random.randint(5,20)
    tavg = random.randint(tmin+1,tmax-1)
    max_t.append(tmax)
    min_t.append(tmin)
    avg_t.append(tavg)
    temp.append([i,tmin,tmax,tavg])
```

```python
df = pd.DataFrame(temp, columns= ["Day","Min","Max","Avg"])
df.to_csv("temp.csv")
df.head()
```

```python
df['target'] = df.apply(lambda row: 1 if row.Max-row.Avg<row.Avg-row.Min else 0 , axis=1)
df.head()
```

```python
x = df[["Min","Max","Avg"]]
y = df["target"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
lgr = LogisticRegression()

lgr.fit(x_train,y_train)
```

```python
lgr.score(x_test,y_test)
```

```python
lgr.score(x_train,y_train)
```

## Chart Plotting

```python
plt.figure(figsize=(12,5))
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(days,min_t,color='blue',label="Min")
plt.plot(days,max_t,color='red',label="Max")
plt.plot(days,avg_t,color='green',label="Average")
plt.legend(loc="best",shadow=True)
plt.grid()
```
